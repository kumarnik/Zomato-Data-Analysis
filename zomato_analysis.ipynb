### Pre-booking a Lunch/Dinner?
#### Here are some of the top restaurants in Delhi & it's neighbouring areas with the following characterstics -

         Have a rating > 4
         Have an option to pre-book a table
         Had at least 100 people rate it on Zomato

```python
top_fine_dine = ncr_data.loc[(ncr_data['Has Table booking'] == "Yes") & (ncr_data['Aggregate rating'] > 4) &
                             (ncr_data['Votes'] > 100)].sort_values('Aggregate rating', ascending = False)
x_ax = top_fine_dine.head(15)['Restaurant Name']
y_ax = top_fine_dine.head(15)['Aggregate rating']


plt.figure(figsize=(15,8))
plt.bar(x_ax,y_ax)
plt.title("Restaurant's rating",fontsize=25)
plt.ylabel("rating",fontsize=20)
plt.xlabel("restaurant",fontsize=20)
plt.xticks(rotation=90)
```

```python
ncr_data.drop(['Restaurant ID','Restaurant Name','Address','Locality Verbose'],axis=1,inplace=True)
```

```python
ncr_data.columns
```

```python
ncr_data.columns=['City', 'Cuisines', 'Average_Cost_for_two', 'Has_Table_booking',
       'Has_Online_delivery', 'Price_range', 'Rating', 'Rating_text',
       'Votes']
```

```python
ncr_data.columns
```

```python
dictionary = {'Excellent': 5,'Very Good': 4,'Average': 2,'Good': 3,'Not rated': 2,'Poor': 1} 
ncr_data.Rating_text = [dictionary[item] for item in ncr_data.Rating_text] 
```

```python
ncr_data['Rating_text'].value_counts()
```

```python
ncr_data.info()
```

```python
Binary= {'Yes': 1,'No': 0} 

ncr_data.Has_Online_delivery = [Binary[item] for item in ncr_data.Has_Online_delivery] 
ncr_data.Has_Table_booking = [Binary[item] for item in ncr_data.Has_Table_booking] 
```

```python
ncr_data.head()
```

```python
city_code={'New Delhi':1,'Noida':2,'Gurgaon':3,'Faridabad':4}
ncr_data.City = [city_code[item] for item in ncr_data.City] 
```

```python
ncr_data.head()
```

```python
ncr_data.drop(['Cuisines'],axis=1,inplace=True)
```

```python
ncr_data.head()
```

```python
ncr_data.shape
```

```python
X=ncr_data.drop(['Rating'],axis=1)
Y=ncr_data['Rating']
```

```python
X.shape
```

#### Train-Test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape , x_test.shape
```

## Modelling:

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_lr=lr.fit(x_train,y_train)
y_pred_lr=model_lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2score_lr=r2_score(y_test,y_pred_lr)
r2score_lr
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_lr))
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test,np.round(y_pred_lr,decimals=1),'*')
plt.xlabel('Actual Rating',size=11)
plt.ylabel('Predicted Rating using Linear Regressor',size=11)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_dt=dt.fit(x_train,y_train)
y_pred_dt=model_dt.predict(x_test)
```

```python
r2score_dt=r2_score(y_test,y_pred_dt)
print("R2 SCORE:",r2score_dt)
print("RMSE value of DecisionTREE is :",np.sqrt(mean_squared_error(y_test,y_pred_dt)))
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test,np.round(y_pred_dt,decimals=1),'*')
plt.xlabel('Actual Rating',size=11)
plt.ylabel('Predicted Rating using Decision Tree',size=11)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=15)
model_rf=rf.fit(x_train,y_train)
y_pred_rf=model_rf.predict(x_test)
```

```python
r2score_rf=r2_score(y_test,y_pred_rf)
print("R2 SCORE:",r2score_rf)
print("RMSE value of Random Forest is :",np.sqrt(mean_squared_error(y_test,y_pred_rf)))
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test,np.round(y_pred_rf,decimals=1),'*')
plt.xlabel('Actual Rating',size=11)
plt.ylabel('Predicted Rating using Random Forest',size=11)
plt.show()
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
model_xgb=xgb.fit(x_train,y_train)
y_pred_xgb=model_xgb.predict(x_test)
```

```python
r2score_xgb=r2_score(y_test,y_pred_xgb)
print("R2 SCORE:",r2score_xgb)
print("RMSE value of XG Boost is :",np.sqrt(mean_squared_error(y_test,y_pred_xgb)))
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_test,np.round(y_pred_xgb,decimals=1),'*')
plt.xlabel('Actual Rating',size=13)
plt.ylabel('Predicted Rating using XG Boost',size=13)
plt.show()
```
